```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

```python
#import 2008 monthly stock data from csv
stock_data_file = "Resources/monthly_2008_data.csv"
monthly_stock_2008 = pd.read_csv(stock_data_file)
monthly_stock_2008.reset_index()
```

```python
x_axis = monthly_stock_2008["Date"]
y_axis1 = monthly_stock_2008["Price"]
```

```python
#import housing data from csv
housing_data_file = "Resources/housing_data.csv"
housing = pd.read_csv(housing_data_file)
housing
```

```python
housing.set_index('date',inplace = True)
housing_2008_data = housing["2006-01-01":"2011-12-01"]
housing_2008_data.reset_index()
```

```python
y_axis2 = housing_2008_data["avg_price"]
```

```python

fig,ax1 = plt.subplots(figsize=(12,6))
ax1.plot(x_axis,y_axis1)
ax1.set_xlabel("Year")
ax1.set_ylabel("Stock Price (USD)")
ax2 = ax1.twinx()
ax2.plot(x_axis,y_axis2,color = "red")
ax2.set_ylabel("Average House Price (CAD)")
plt.title("Average Monthly Stock Price  Vs.Average Monthly House Price from 2006 - 2011")
ax1.xaxis.set_major_locator(plt.MaxNLocator(71))
ax1.set_xticklabels(x_axis,rotation="vertical")
plt.xlim(-0.05, len(x_axis)-1)
ax1.axvspan(34.5, 52.5, alpha=0.5, color='yellow')

plt.tight_layout()
```
