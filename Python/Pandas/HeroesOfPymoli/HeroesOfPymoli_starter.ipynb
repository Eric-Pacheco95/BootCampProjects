```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Get Unique Players into list and find length of list to get total number of unique players
player_list = purchase_data['SN'].unique()
player_list_length = len(player_list)
#Make new dataframe with Total Players as the column 
player_list_df = pd.DataFrame({'Total Players':[player_list_length]})
player_list_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Make a list of unique items then find length
UniqueItems = purchase_data['Item Name'].unique()
Num_UniqueItems = len(UniqueItems)

#Get Average Price, Number of Purchases and Total revenue using purchase_data 
AveragePrice = purchase_data['Price'].mean()
NumberOfPurchases = purchase_data['Purchase ID'].count()
TotalRevenue = purchase_data['Price'].sum()

#Create Summary Dataframe then format Average Price and Total Revenue columns
summary_df = pd.DataFrame({'Number Of Unique Items':[Num_UniqueItems],'Average Price':[AveragePrice],
                           'Number Of Purchases':[NumberOfPurchases],'Total Revenue':[TotalRevenue]})
summary_df['Average Price'] = summary_df['Average Price'].map("${:.2f}".format) 
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:.2f}".format) 
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Make new dataframe with all unique players and their genders
unique_players_df = purchase_data.drop_duplicates(['SN'])
unique_players_df.sort_values('SN',inplace=True)
unique_players_df.set_index('SN',inplace=True)
gender_counts = unique_players_df['Gender'].value_counts()

#Make new Dataframe with Total Count and Percentage of Players based on Gender, then format percentage column
gender_counts_df = pd.DataFrame({'Total Count':gender_counts,'Percentage Of Players':gender_counts/player_list_length*100})
gender_counts_df['Percentage Of Players'] = gender_counts_df['Percentage Of Players'].map("{:.2f}%".format)
gender_counts_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Make new DataFrame Based on Gender
purchase_data_genders = purchase_data.groupby(['Gender'])
#Find values for new dataframe
Total_Purchase_Value = purchase_data_genders['Price'].sum()
Average_Purchase_Price = purchase_data_genders['Price'].mean()
Purchase_Count = purchase_data['Gender'].value_counts()
#Make new DataFrame based on unique players
unique_players_groupby = purchase_data.groupby(['SN'])
unique_players_df['Average Purchase'] = unique_players_groupby['Price'].sum()
gender_groupby_averageprice = unique_players_df.groupby(['Gender'])
Average_Total_Purchase_Per_Person = gender_groupby_averageprice['Average Purchase'].mean()
Average_Total_Purchase_Per_Person = Average_Total_Purchase_Per_Person.map("${:.2f}".format)
#Make new summary dataframe
Purchasing_Analysis_df = pd.DataFrame({'Purchase Count':Purchase_Count,'Average Purchase Price':Average_Purchase_Price,
                                      'Total Purchase Value':Total_Purchase_Value,'Average Total Purchase Per Person'
                                       :Average_Total_Purchase_Per_Person})
Purchasing_Analysis_df['Average Purchase Price'] = Purchasing_Analysis_df['Average Purchase Price'].map("${:.2f}".format)
Purchasing_Analysis_df['Total Purchase Value'] = Purchasing_Analysis_df['Total Purchase Value'].map("${:.2f}".format)
Purchasing_Analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Make bins and labels for each bin
bins = [0,9,14,19,24,29,34,39,100]
ranges = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

#Place each age into their respective bins and find the count of each age range
Total_Count = pd.cut(unique_players_df['Age'],bins,labels=ranges)
age_analysis_df = pd.DataFrame({'Age Range':Total_Count})
Ages_Ranges_Count = age_analysis_df['Age Range'].value_counts(sort=False)

#Make age demographics table displaying total count and percentage of players for each age range
Age_demographics_df = pd.DataFrame({'Total Count':Ages_Ranges_Count,'Percentage Of Players':Ages_Ranges_Count/player_list_length*100})
Age_demographics_df['Percentage Of Players'] = Age_demographics_df['Percentage Of Players'].map("{:.2f}%".format)
Age_demographics_df
```

## Top Spenders

```python
#Use bins in above cell to add age range data to copy of original purchase_data
purchase_data_binned = purchase_data.copy()
purchase_data_binned['Age Range'] = pd.cut(purchase_data_binned['Age'],bins,labels=ranges)

#Group by age ranges and get Purchase Count, Average Purchase Price, and Total Purchase Value for each age range, then format 
age_range_groupby = purchase_data_binned.groupby(['Age Range'])
Purchase_Count_Agerange = age_range_groupby['Age Range'].count()
Average_Purchase_Price_Agerange = age_range_groupby['Price'].mean()
Average_Purchase_Price_Agerange = Average_Purchase_Price_Agerange.map("${:.2f}".format)
Total_Purchase_Value_Agerange = age_range_groupby['Price'].sum()
Total_Purchase_Value_Agerange = Total_Purchase_Value_Agerange.map("${:.2f}".format)

#Get Avg Total Purchase per Person
purchase_data_AgeRanges = purchase_data.copy()
purchase_data_AgeRanges['Age Range'] = pd.cut(purchase_data_binned['Age'],bins,labels=ranges)
purchase_data_AgeRanges_grouped = purchase_data_AgeRanges.groupby(['SN'])
Age_Ranges = purchase_data_AgeRanges_grouped['Age Range'].first()
Total_purchases_PerPerson = purchase_data_AgeRanges_grouped['Price'].sum()
Total_purchases_WithAgeRange = pd.DataFrame({'Age Range':Age_Ranges,'Total Purchases':Total_purchases_PerPerson})
Total_purchases_WithAgeRange_grouped = Total_purchases_WithAgeRange.groupby(['Age Range'])
Avg_Total_Purchase_per_Person = Total_purchases_WithAgeRange_grouped['Total Purchases'].mean()
Avg_Total_Purchase_per_Person = Avg_Total_Purchase_per_Person.map("${:.2f}".format)

#Make Summary DataFrame with the results above
age_range_summary = pd.DataFrame({'Purchase Count':Purchase_Count_Agerange,'Average Purchase Price':Average_Purchase_Price_Agerange,
                                  'Total Purchase Value': Total_Purchase_Value_Agerange})
age_range_summary['Average Total Purchase Per Person'] = Avg_Total_Purchase_per_Person
age_range_summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Make new dataframe with Item ID, Item Name and Item Price
items_df = purchase_data[['Item ID','Item Name','Price']]
items_df_grouped = items_df.groupby(['Item ID','Item Name'])
Item_Purchase_Count = items_df_grouped['Item ID'].count()
Item_Price = items_df_grouped['Price'].first()
Total_Item_Purchase_Value = items_df_grouped['Price'].sum()

#Make new dataframe using above data
Item_Analysis_df = pd.DataFrame({'Purchase Count':Item_Purchase_Count,'Item Price':Item_Price,'Total Purchase Value':Total_Item_Purchase_Value})

#Copy above dataframe, sort values by purchase count, then format columns
Formatted_Item_Analysis_df = Item_Analysis_df.copy()
Formatted_Item_Analysis_df.sort_values('Purchase Count',inplace=True,ascending=False)
Formatted_Item_Analysis_df['Total Purchase Value'] = Formatted_Item_Analysis_df['Total Purchase Value'].map("${:.2f}".format)
Formatted_Item_Analysis_df['Item Price'] = Formatted_Item_Analysis_df['Item Price'].map("${:.2f}".format)
Formatted_Item_Analysis_df.head()
```
